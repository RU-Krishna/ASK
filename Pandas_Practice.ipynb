```python
import pandas as pd
import numpy as np
import calendar as cd
```

```python
# EngAlphabet

engAlpha = pd.Series(['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z'],
                     index = range(1,27))
print(engAlpha)

# Vowels as index label and their values set to 0.

vowels = pd.Series(np.zeros(5, dtype = int), index = ['a','e','i','o','u'])
print(f'Vowels: \n{vowels}')

# Your 5 friends from dictionary having name as the key and roll no as the value.
# data is given in dictionary form.
friends = pd.Series(data = {
    'Krishna': 2150080580020,
    'Rds': 2150080580042,
    'Swakshat': 2150080580054,
    'Trisha': 2150080580056,
    'Pray': 2150080580035
})
print(f'Friends: \n{friends}')

# Month Days
months = list(cd.month_name)
del months[0]
monthDays = pd.Series(
    data = pd.array([31,29,31,30,31,30,31,31,30,31,30,31]),
    index = months
)
print(f'Month Days:\n{monthDays}')
```

```python
# set all the values in the vowel series to 10

vowels.replace(to_replace = 0, value = 10, inplace= True)
print(f'Vowels: \n{vowels}')

vowels = vowels.divide(2)
print(f'After Dividing by 2:\n{vowels}')


vowel1 = pd.Series(data = [2,5,6,3,8], index = ['a','e','i','o','u'])
print(f'New vowel Series: \n{vowel1}')

vowel3 = vowels + vowel1
print(f'New Vowel Series: \n{vowel3}')


vowel_sub = vowels.sub(vowel1)
vowel_mul = vowels.mul(vowel1)
vowel_div = vowels.div(vowel1)

print(f'Vowel  Subtraction: \n{vowel_sub}')
print(f'Vowel Multiplication: \n{vowel_mul}')
print(f'Vowel Divide: \n{vowel_div}')

# Altering the labels of vowels to capital ones.
result = vowels.rename(index = {'a':'A', 'e': 'E', 'i':'I', 'o':'O', 'u':'U'})
print(f'After replacing labels: \n{result}')


```

```python

sales_person = ['Madhu', 'Kusum', 'Kinshuk', 'Ankit', 'Shruti']
sales_data = {
    '2014': pd.Series([100.5, 150.5, 200.9,30000,40000], index = sales_person),
    '2015': pd.Series([12000,18000,22000,30000,45000], index = sales_person),
    '2016': pd.Series([20000,50000,70000,100000,12500], index = sales_person),
    '2017': pd.Series([50000,60000,70000,80000,90000], index = sales_person)
}


salesDF = pd.DataFrame(
    data = sales_data
)
print(f'Sales Data: \n{salesDF}')

#Display the row labels of the sales Data FRame
print(f'Row Labels: \n{salesDF.index}')

#Display the column labels of the Data Frame
print(f'Column Labels: \n{salesDF.columns}')

# Display the dimesnions, shape, size and values of the Data FRame.
print(f'Dimensions: {salesDF.ndim} \nShape: {salesDF.shape} \nSize: {salesDF.size} \nValues: {salesDF.values}')

#Display the last two rows of the DF.
print(f'Last 2 rows of te Data Frame: \n{salesDF.iloc[0: 2, 0: ]}')

#Display the first two column
print(f'First two columns of the Data Frame: \n{salesDF.iloc[0: , 2: ]}')


#Sales 2 of 2018
sales2_dict = {
    '2018': pd.Series(data = [160000,110000,500000,340000,900000], index = sales_person)
}
sales2DF = pd.DataFrame(sales2_dict)
print(f'2018 sales: \n{sales2DF}')


#Check if sales 2 is empty or not
print(f'Is sales2 empty: \n{sales2DF.empty}')
```

```python
newSalesDF = pd.concat(objs = [salesDF, sales2DF], axis = 1)
print(f'New Sales DF: \n{newSalesDF}')

#Transpose he new Sales Data Frame
print(f'Transpose of the new Sales Data Frame: \n{newSalesDF.T}')

#Display the sales made by each person in 2017
print(f"Sales in 2017: \n{newSalesDF.loc[ :,'2017']}")

#Sales by madhu and ankit in 2017 & 2018
print(f'Sales by Ankit & Madhu in 2017 & 2018: \n{newSalesDF.loc[["Ankit", "Madhu"], ["2017", "2018"]]}')

#Sales made by Shruti in 2016
print(f'Sales by Shruti in 2016: \n{newSalesDF.loc["Shruti", "2016"]}')

#Add data for salesman Summet for all the year.
newSalesDF.loc['Sumeet'] = [196.2, 38700, 52000, 78438,38852]
print(f'New sales after adding Sumeet: \n{newSalesDF}')

# after dropping sales of 2014
sales2014 = newSalesDF.drop('2014', axis = 1)
print(f'After deletion: \n{sales2014}')

# delete sales for Kinushk
kinushkRemove = newSalesDF.drop('Kinshuk', axis = 0)
print(f'After kinsushk removal: \n{kinushkRemove}')

# Chnage the name
renameDF = newSalesDF.rename({'Ankit': 'Vivaan', 'Madhu': 'Shailesh'}, axis= 0)
print(f'After renaming: \n{renameDF}')

#Sale value updation
renameDF.loc['Shailesh', '2018'] = 100000
print(f'Shailesh Sales Updation for 2018: \n{renameDF}')
```
